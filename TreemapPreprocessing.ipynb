Data Formatting for Treemap Visualizations

```python
import json 
```

Area Treemap

```python
Treemap1={
    'name' : "Root",
    'children': [
        {
            'name' : "North LA",
            'children' : []
        },
        {
            'name': "South LA",
            'children' : [] 
        }
    ]
}
North_LA=['N Hollywood','Foothill','Mission','Devonshire','Topanga','West Valley','Van Nuys']
South_LA=[x for x in df['AREA NAME'].unique() if x not in North_LA]

for val in North_LA:
    counter=df['AREA NAME'].value_counts()[val]
    Treemap1['children'][0]['children'].append({
        'name' : val,
        'value' : str(counter)
    })

for val in South_LA:
    counter=df['AREA NAME'].value_counts()[val]
    Treemap1['children'][1]['children'].append({
        'name' : val,
        'value' : str(counter)
    })

Treemap1
```

```python
with open("Treemap1.json", "w") as outfile: 
    json.dump(Treemap1, outfile)
```

Date of Occurance Treemap

```python
Treemap2={
    "name" : "Root",
    "children" : []
}
for x in ['2020','2021','2022','2023']:
    Treemap2['children'].append({
        'name': x,
        'children' : []
    })

for p,y in enumerate(['2020','2021','2022','2023']):
    Months=[]
    Conversion={}
    for i in range(1,13):
        if(i<10):
            Months.append({'name' : '0'+str(i), 'value' : '0'})
        else :
            Months.append({'name': str(i), 'value' : '0'})

    for x in df['DATE OCC']:
        spliting=x.split('/')
        if(y in spliting[2]):
            for x in Months:
                if x['name'] == spliting[0]:
                    x['value']=str(int(x['value'])+1)
    
    Conversion['01']='January'
    Conversion['02']='February'
    Conversion['03']='March'
    Conversion['04']='April'
    Conversion['05']='May'
    Conversion['06']='June'
    Conversion['07']='July'
    Conversion['08']='August'
    Conversion['09']='September'
    Conversion['10']='October'
    Conversion['11']='November'
    Conversion['12']='December'

    for x in Months:
        x['name']=Conversion[x['name']]
    Treemap2['children'][p]['children']=Months

Treemap2
```

```python
with open("Treemap2.json", "w") as outfile: 
    json.dump(Treemap2, outfile)
```
